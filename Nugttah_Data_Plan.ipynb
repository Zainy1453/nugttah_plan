### Nugttahs's data vision

As a data professional, I know that building the data capacity of Nugttah to achieve the goal of being a data-driven company requires careful planning and execution. 

For that goal I have developed a vision that is broken down into three major periods with three major themes that will span a period of one year. For example, the first period will be focused on Data Backend, with the primary allocation of resources and talent to this focus, while moving at a slower pace in the two other themes and the same goes for the periods that follow.

I understand that planning is not a linear process, but rather an iterative one, and that adversity is always the best teacher. With this in mind, I am prepared to adapt and adjust as necessary to ensure the successful implementation of the gist of this plan and to realzie the goals it promises.


#### Theme: Data Backend (2-3 months)

The data backend theme is aimed at enabling Nugttah to assess the current state of its data landscape, establish a comprehensive data strategy in collaboration with the business team, build a robust data infrastructure and governance framework, and develop an actionable data plan with clear OKRs for the data team. This theme will set the stage for Nugttah's data journey and will be instrumental in establishing a solid foundation to support the company's data goals. By the end of this theme, Nugttah will have a well-defined data infrastructure, a clear roadmap for data collection, storage, and usage, and a framework for data management, privacy, and security that is aligned with business goals.

<br>

**Objective 1: Assess the current company data landscape**

To achieve this objective, the data team will:

- **Engage in Nugttah's day-to-day business/data process to model our current process, data flow model, and data infrastructure.**

> During this process, the data team will use process mapping techniques to gain a visual understanding of the company's data flow, infrastructure, and sources. This will include examining the current data sources and fact tables being used, as well as assessing the data quality and relations between different data sources. By engaging with the day-to-day business operations, the data team will be able to understand how data is being used, which will inform the development of a comprehensive data strategy.

> Overall, this objective is focused on providing a basic understanding of the company's data status quo, which will help guide the development of the data plan. It will also serve as a foundation for the development of the company's overall data strategy.

**Objective 2: Define a comprehensive data strategy with business**

To achieve this objective, the data team will:

- *****Define business goal(s) and the organizational north star.*****
> This involves engaging in conversations with the business to identify and understand the current business goals, and how they align with the company's overall vision and mission. This step serves a cultural purpose as it allows the data team to immerse themselves in the business and feed their bias towards business acumen and relevance. This is a crucial step as it helps bridge the gap between the data team and the rest of the company and aligns their efforts towards a common goal. 

- *****Define the data goals, while considering the nature of the organization and what is strategic and tactical.*****
> This involves defining the specific data goals that will support the business goals, taking into account the company's vision and mission, and identifying what data is needed to achieve those goals. This requires a balance between long-term strategic planning and short-term business considerations.

- *****Outline a clear roadmap using OKRs for data collection, storage, and usage that can help the company achieve its data goals and business goals.*****

> Here we will create a clear and actionable plan for data collection, storage, and usage, which will serve as a roadmap for the data team to follow. This roadmap will be based on the defined data goals and will include specific objectives and key results (OKRs) that will enable the company to track progress towards achieving its data goals and business goals.



**Sample of OKRs:**


> **<small> `OKRs` is a project management activity that outlines the company plan and the each department's plan which is set with specific criteria and serves as a viable actionable timed goals for the business overall and the its departments contribution to those goals</small>**

**Objective 3: Develop a Robust Data Infrastructure for Implementing the Data Roadmap with Ease**

To achieve this objective, the data team will:

- *****Evaluate the current data infrastructure, including the hardware and software used for data collection, storage, processing, and analysis.*****

>This will involve assessing the Postgres data warehouse, data integration tools, and data pipeline tools currently in use.
By conducting a thorough evaluation of the current data infrastructure, the data team will be able to identify any areas that require improvement and develop a plan to address them. This will ensure that the infrastructure is robust and can support the implementation of the data roadmap. The team will also focus on proper mapping of our Mongo DB with a SQL-based warehouse, and ensure it has an analysis-friendly schema with reliable connectivity to our BI tool.

- *****Evaluate the quality of data from different sources by assessing its completeness, accuracy, consistency, and timeliness.*****

> The data team will use data profiling techniques to identify any anomalies or inconsistencies that may require data cleansing or standardization.By ensuring the quality of data from different sources, the team can avoid any potential data silos or redundancies that may affect data quality. To achieve this, the team will establish a viable data validity framework. This framework will also address the question of how far back the team can go to ensure the validity of Nugttah's data points, and identify any possible data pools that may lack integrity. This step is crucial to building a strong data infrastructure that supports the data roadmap with ease of use and quality data points to achieve our data goals.



**Objective 4: Establish a robust data governance framework that addresses data management, privacy, and security.**
To achieve this, the following steps will be taken:


- *****Develop procedures for privacy and security that comply with relevant data protection laws and regulations.*****

>To ensure that sensitive data is protected and that all relevant laws and regulations are complied with, the data team will establish procedures and guidelines for privacy and security

- *****Identify data access levels (public, confidential, personal data, business data) and develop a framework for access control.*****

> The data team will determine the various data access levels that exist within the organization and develop a permissions framework with the data environment to control access to data, based on these levels.

- *****Establish NDAs and other agreements to protect confidential information.*****

>To ensure that confidential information is protected, we will establish nondisclosure agreements and other agreements that protect the confidentiality of sensitive data.

Organizational health metrics:

Employee engagement: Percentage of employees who report feeling engaged with their work, as measured by an employee engagement survey. For example, if a survey finds that 75% of employees report feeling engaged, this would be the engagement metric.

Turnover rate: Percentage of employees who leave the company within a given time period, such as a year. For example, if a company has 100 employees and 10 leave within a year, the turnover rate would be 10%.
Organizational culture: Score on a survey that measures the alignment of the company's culture with its values and goals. For example, if a culture survey gives the company a score of 80 out of 100, this would be the culture metric.

HR metrics:

Time-to-hire: Average number of days it takes to fill a job opening. For example, if it takes an average of 30 days to fill open positions, this would be the time-to-hire metric.

Retention rate: Percentage of employees who stay with the company for a certain period of time, such as a year. For example, if a company has 100 employees and 90 remain after a year, the retention rate would be 90%.

Employee satisfaction: Score on a survey that measures employee satisfaction with various aspects of their job, such as compensation and benefits, work-life balance, and career growth opportunities. For example, if an employee satisfaction survey gives the company a score of 4 out of 5, this would be the satisfaction metric.
Performance metrics:

Productivity: Measure of the amount of work accomplished by an individual or team in a given time period. For example, if an employee is able to complete 5 tasks per hour, this would be the productivity metric.

Quality of work: Measure of the accuracy and effectiveness of an individual or team's work. For example, if a team completes 100 projects with a 90% accuracy rate, this would be the quality metric.

Customer satisfaction: Score on a survey that measures customer satisfaction with a company's product or service. For example, if a customer satisfaction survey gives the company a score of 8 out of 10, this would be the satisfaction metric.
